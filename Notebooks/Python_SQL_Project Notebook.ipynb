Filling misisng Data

```python
# Order Dataset
orders['order_approved_at'] = orders['order_approved_at'].fillna(orders['order_purchase_timestamp'])

# Leave delivery dates as NaT (we may want to ignore incomplete orders in some analysis)
orders['order_delivered_carrier_date'] = pd.to_datetime(orders['order_delivered_carrier_date'], errors='coerce')
orders['order_delivered_customer_date'] = pd.to_datetime(orders['order_delivered_customer_date'], errors='coerce')
```

```python
# Product Dataset
# Fill missing category with 'Unknown'
products['product category'] = products['product category'].fillna('Unknown')

# Fill missing lengths/descriptions/photos with 0
products['product_name_length'] = products['product_name_length'].fillna(0)
products['product_description_length'] = products['product_description_length'].fillna(0)
products['product_photos_qty'] = products['product_photos_qty'].fillna(0)

# Fill small numeric gaps with median
products['product_weight_g'] = products['product_weight_g'].fillna(products['product_weight_g'].median())
products['product_length_cm'] = products['product_length_cm'].fillna(products['product_length_cm'].median())
products['product_height_cm'] = products['product_height_cm'].fillna(products['product_height_cm'].median())
products['product_width_cm'] = products['product_width_cm'].fillna(products['product_width_cm'].median())
```

# Merge of the data

```python
# Start with order_items as main table
master_df = order_items.merge(orders, on='order_id', how='left')\
                       .merge(products, on='product_id', how='left')\
                       .merge(sellers, on='seller_id', how='left')\
                       .merge(customers, on='customer_id', how='left')\
                       .merge(payments, on='order_id', how='left')
```

```python
# Get unique zip codes in master table
customer_zips = master_df['customer_zip_code_prefix'].unique()
seller_zips = master_df['seller_zip_code_prefix'].unique()
```

```python
# Customer geolocation: only keep one row per zip code
customer_geo_unique = geolocation.drop_duplicates(subset=['geolocation_zip_code_prefix'])

# Seller geolocation: same
seller_geo_unique = geolocation.drop_duplicates(subset=['geolocation_zip_code_prefix'])
```

```python
# Customer geolocation mapping
customer_geo_dict = customer_geo_unique.set_index('geolocation_zip_code_prefix')[
    ['geolocation_lat', 'geolocation_lng', 'geolocation_city', 'geolocation_state']
].to_dict('index')

# Seller geolocation mapping
seller_geo_dict = seller_geo_unique.set_index('geolocation_zip_code_prefix')[
    ['geolocation_lat', 'geolocation_lng', 'geolocation_city', 'geolocation_state']
].to_dict('index')
```

```python
# Customer
master_df['customer_lat'] = master_df['customer_zip_code_prefix'].map(lambda x: customer_geo_dict.get(x, {}).get('geolocation_lat'))
master_df['customer_lng'] = master_df['customer_zip_code_prefix'].map(lambda x: customer_geo_dict.get(x, {}).get('geolocation_lng'))
master_df['customer_city_geo'] = master_df['customer_zip_code_prefix'].map(lambda x: customer_geo_dict.get(x, {}).get('geolocation_city'))
master_df['customer_state_geo'] = master_df['customer_zip_code_prefix'].map(lambda x: customer_geo_dict.get(x, {}).get('geolocation_state'))

# Seller
master_df['seller_lat'] = master_df['seller_zip_code_prefix'].map(lambda x: seller_geo_dict.get(x, {}).get('geolocation_lat'))
master_df['seller_lng'] = master_df['seller_zip_code_prefix'].map(lambda x: seller_geo_dict.get(x, {}).get('geolocation_lng'))
master_df['seller_city_geo'] = master_df['seller_zip_code_prefix'].map(lambda x: seller_geo_dict.get(x, {}).get('geolocation_city'))
master_df['seller_state_geo'] = master_df['seller_zip_code_prefix'].map(lambda x: seller_geo_dict.get(x, {}).get('geolocation_state'))
```

```python
master_df.to_csv('master_olist_clean.csv', index=False)
```
